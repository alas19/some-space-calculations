```python
import pandas as pd
import math 
from matplotlib import pyplot as plt
from scipy import stats
import numpy as np
```

```python
fgm = pd.read_csv(r'C:\Users\fener\Desktop\THC_L2_FGM_46034.csv')
```

```python
mom = pd.read_csv(r'C:\Users\fener\Desktop\THC_L2_MOM_46034.csv')
```

```python
fgm_new = pd.DataFrame()
for i in range(len(fgm)):
    if i % 20 == 0:
        fgm_new = fgm_new.append( fgm.iloc[i,:])
fgm_new = fgm_new.reset_index()
```

```python
fgm_new.drop(fgm_new.tail(2).index,inplace=True)
```

```python
fgm_new
```

```python
fgm_new.describe()
```

```python
mom_new = pd.DataFrame()
for i in range(len(mom)):
    if i % 20 == 0:
        mom_new = mom_new.append( mom.iloc[i,:])
mom_new = mom_new.reset_index()
```

```python
mom_new
```

```python
velocity = [1]*76
for i in range(len(mom_new)):
    vtot = mom_new.iloc[i,3]**2 + mom_new.iloc[i,4]**2 + mom_new.iloc[i,5]**2
    vsqrt = vtot**(1/2)
    velocity[i] = vsqrt
```

```python
velocity
```

```python
p_dyn = [1]*76
for i in range(len(velocity)):
    p_dyn[i] = (velocity[i]**2)*mom_new.iloc[i,2]
```

```python
p_dyn
```

```python
stats.describe(p_dyn)
```

```python
Rmp = [1]*76
for i in range(len(Rmp)):
    Rmp[i] = (2*(fgm_new.iloc[i,5]**2)/((4*math.pi*(10**-7))*p_dyn[i]))**(1/6)
```

```python
Rmp
```

```python
stats.describe(Rmp)
```
